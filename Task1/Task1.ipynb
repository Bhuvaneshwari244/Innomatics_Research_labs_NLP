### Task 1: Conceptual Understanding

**1. What is the difference between "Love" and "love" in NLP?**
In standard NLP processing, "Love" (capitalized) and "love" (lowercase) are treated as two distinct tokens because computers read their underlying ASCII/Unicode values differently. Converting text to lowercase normalizes the data.

**2. What happens if stopwords are not removed?**
If stopwords (like "is", "the", "and") are not removed, they dominate the frequency distribution, increasing data dimensionality and making it harder for algorithms to identify truly meaningful words.

**3. Provide two real-world scenarios where removing stopwords can be harmful.**
* **Sentiment Analysis:** Removing words like "not" or "no" can flip the meaning (e.g., "I am not happy" becomes "I am happy").
* **Machine Translation:** Stopwords are essential for grammatical structure; removing them results in broken translations.

**4. Explain the difference between stemming and lemmatization.**
* **Stemming:** A heuristic approach that chops off word ends to find the root (e.g., "running" -> "run"). It is fast but can create non-dictionary words.
* **Lemmatization:** A morphological analysis that reduces a word to its meaningful dictionary base form (e.g., "better" -> "good").


# Tasks 2 & 3 (Preprocessing & Stress Testing)

```python
import re

# Task 2: Advanced Preprocessing Function
def preprocess_text(text):
    if not isinstance(text, str) or not text.strip():
        return [], ""
    
    text = re.sub(r'http\S+|www\.\S+', '', text) # Remove URLs
    text = re.sub(r'\S+@\S+', '', text)          # Remove emails
    text = text.lower()                          # Lowercase
    text = re.sub(r'\d+', '', text)              # Remove numbers
    text = re.sub(r'[^\w\s]', ' ', text)         # Remove punctuation/emojis
    text = re.sub(r'(.)\1{2,}', r'\1', text)     # Handle repeated characters
    
    tokens = text.split()
    exceptions = {"no", "not", "ok"} 
    cleaned_tokens = [word for word in tokens if len(word) > 2 or word in exceptions]
    cleaned_sentence = " ".join(cleaned_tokens)
    
    return cleaned_tokens, cleaned_sentence

# Task 3: Stress Testing
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("--- TASK 3: STRESS TESTING RESULTS ---")
for text in sample_inputs:
    tokens, cleaned_text = preprocess_text(text)
    print(f"Original Text    : {text}")
    print(f"Cleaned Tokens   : {tokens}")
    print(f"Cleaned Sentence : {cleaned_text}\n")
```

# Task 4 (Token Analytics)

```python
# Task 4: Token Analytics
print("--- TASK 4: TOKEN ANALYTICS ---")
for text in sample_inputs:
    tokens, _ = preprocess_text(text)
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0
    
    print(f"Text: '{text}'")
    print(f"  -> Total Tokens: {total_tokens}")
    print(f"  -> Unique Tokens: {unique_tokens}")
    print(f"  -> Avg Token Length: {avg_len:.2f}\n")
```

# Tasks 5, 6 & 7 (Frequency, Pipeline, Error Handling)

```python
from collections import Counter

# Task 6: Build Full Pipeline
def full_pipeline(text_list):
    all_tokens, all_clean_sentences = [], []
    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        if clean_sentence:
            all_clean_sentences.append(clean_sentence)
    return {"tokens": all_tokens, "clean_sentences": all_clean_sentences}

pipeline_results = full_pipeline(sample_inputs)

# Task 5: Frequency Analysis
print("--- TASK 5: FREQUENCY ANALYSIS ---")
token_counts = Counter(pipeline_results["tokens"])
print("Top 10 Most Frequent:", token_counts.most_common(10))
print("Top 5 Least Frequent:", token_counts.most_common()[-5:])

# Task 7: Error Handling
print("\n--- TASK 7: ERROR HANDLING ---")
error_tests = ["", "😊😂🚀", "12345 67890", "    "]
for text in error_tests:
    tokens, clean_sentence = preprocess_text(text)
    print(f"Input: '{text}' -> Tokens: {tokens} | Cleaned: '{clean_sentence}'")
```
